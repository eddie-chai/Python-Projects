# Task 2 - Visualising the relationship between Annual Leave available and bonus

```python
#importing numpy, pandas and matplotlib library modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

#importing CSV onto python
df = pd.read_csv(r'Task 2.csv',index_col= False)

#defining x as 'Annual_leave_taken' column and y as 'Bonus' column
x = df['Annual_leave_taken']
y = df['Bonus']

#size of plot
np_x = np.array(x)
np_x = np_x * 5
```

```python
#changing plot style
plt.style.use('seaborn') 

#plotting scatter graph, changing marker size and transparency
plt.scatter(x,y, s = np_x, alpha = 0.5)

#plotting title and x and y-axis labels
plt.title('Annual Leave & Bonus')
plt.xlabel('Annual Leave')
plt.ylabel('Bonus')
plt.scatter(x,y, color ='darkgreen')

#creating texts on graph
plt.text(30, 6500,'Highest Bonus')
plt.scatter(x,y, marker= "o")
plt.text(16, 6500,'R = ' + str(round(x.corr(y),2)))

plt.show()

#calculating correlation
x.corr(y)
```
