# Task 3 - Visualising the revenue by country

```python
#importing pandas and matplotlib library modules
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#importing CSV onto python
df = pd.read_csv(r"C:\Users\cheny\Desktop\1. Generation Project\CSV\Question 3.csv")

#defining x as 'Country' column and y as 'Total Revenue' column
#also rounded total revenue to two decimal places
CountryRegionCode =  df['Country']
Revenue = df['Total Revenue'].round(2)
```

```python
#setting the subplots
fig, ax = plt.subplots(1, 1)

#plotting bar chart with labels
plt.bar(CountryRegionCode,  Revenue)
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title ('Total Revenue By Country')

#adjusting width and height of bar chart
fig.set_figwidth(10)
fig.set_figheight(5)

plt.show()
```
