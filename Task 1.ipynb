# Task 1 - Visualising the regions with the best revenue

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the CSV file onto Pandas
df = pd.read_csv(r'Task 1.csv')
```

```python
SalesYTD = df['SalesYTD']
Region = df['Region']
SalesYTD
```

```python
# Plot the figure.
plt.figure(figsize=(14, 5))
fig = SalesYTD.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('SalesYTD')
plt.title('Region in USA v SalesYTD')
fig.set_xticklabels(Region, rotation = 0)
fig.bar_label(fig.containers[0], label_type='edge')
```
