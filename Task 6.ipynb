# Task 6 - Visualising the relationship between store size, Number of employees, and Total Revenue

```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the CSV file onto Pandas
df = pd.read_csv(r'Question 6.csv')
```

```python
#Assigning the right columns from DataFrames to variables
SquareFeet = df['SquareFeet']
NumberEmployees = df['NumberEmployees']
TotalRevenue = df['Total Revenue']
```

```python
#setting the subplots
fig = plt.figure() 
# Figure size
#fig.set_size_inches(25,10)
fig.set_size_inches(20, 8)
g = sns.lmplot(x = 'SquareFeet', y = 'NumberEmployees',data= df, hue='Total Revenue', fit_reg=False, scatter_kws={"s": 10})
g.fig.set_figwidth(12)
g.fig.set_figheight(10)

#plt.xlabel('SquareFeet')
#plt.ylabel('NumberEmployees')
#plt.title('SquareFeet v NumberEmployees v Total Revenue')
#plt.scatter(SquareFeet, NumberEmployees, s=TotalRevenue*0.00003)

#adjusting the spacing and size of the plot
z1 = np.polyfit(SquareFeet, NumberEmployees, 1)
p1 = np.poly1d(z1)
plt.plot(SquareFeet, p1(SquareFeet))

#plt.tight_layout()
plt.show()
```
