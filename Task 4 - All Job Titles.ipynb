# Task 4 - Visualising the Average Sick Leave by Job Title

```python
#import matplotlib and pandas module libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv (r'Task 4.csv')
```

```python
#filtering organization level 0-2 for DataFrames
#convert it to list for color customization
org_012 = df.loc[df['OrganizationLevel'] <= 2]
list_org = list(org_012['OrganizationLevel'])

#filtering organization level 3 & 4 for DataFrames and sorting values by AvgSickLeave
org3_df = df[df["OrganizationLevel"].isin([3])]
org3_df_sorted = org3_df.sort_values(by=['AverageSickLeave'])

org4_df = df[df["OrganizationLevel"].isin([4])]
org4_df_sorted = org4_df.sort_values(by=['AverageSickLeave'])
```

```python
#creating the empty subplots
plt.style.use('ggplot')
fig, ax = plt.subplots(3, 1)

#plotting the first barcharts for Organisation level 0-2
plt.subplot(3, 1, 1)
df_level_0 = df.loc[df['OrganizationLevel']  < 3]
x = df_level_0['JobTitle']
y = df_level_0['AverageSickLeave']

col = []
for val in list_org:
    if val < 1:
        col.append('tab:blue')
    elif val < 2:
        col.append('tab:orange')
    else:
        col.append('tab:green')

plt.xlabel('Average Sick Leave')
plt.ylabel('Job Title')
plt.title('Avg Sick Leave for Org Lvl 0-2')
l1 = plt.barh(x, y, color='tab:blue')
l2 = plt.barh(x, y, color='tab:orange')
l3 = plt.barh(x, y, color='tab:green')
plt.barh(x, y, color=col)

#creating legends and setting x-axis limit to 80
plt.legend(([l1,l2,l3]), ['CEO', 'Level 1', 'Level 2'], bbox_to_anchor =(1.2, 0.5))
plt.xlim([0, 80])

#inverting figure 1
ax = plt.gca()
ax.invert_yaxis()

#plotting the second barchart for Organisation level 3
plt.subplot(3, 1, 2)
x3 = org3_df_sorted['JobTitle']
y3 = org3_df_sorted['AverageSickLeave']
plt.xlabel('Average Sick Leave')
plt.ylabel('Job Title')
plt.title('Avg Sick Leave for Org Lvl 3')
plt.barh(x3, y3, color='darkslateblue')

#setting x-axis limit to 80
plt.xlim([0, 80])

#plotting the third barchart for Organisation level 4
plt.subplot(3, 1, 3)
x4 = org4_df_sorted['JobTitle']
y4 = org4_df_sorted['AverageSickLeave']
plt.xlabel('Average Sick Leave')
plt.ylabel('Job Title')
plt.title('Avg Sick Leave for Org Lvl 4')
plt.barh(x4, y4, color = 'teal')

#setting x-axis limit to 80
plt.xlim([0, 80])

#adjusting height and width of bar charts
fig.set_figheight(25)
fig.set_figwidth(10)

plt.show()
```
