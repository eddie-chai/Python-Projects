# Task 5 - Visualising the relationship bettwen Store Trading Duration and Annual Revenue

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the CSV file onto Pandas
df = pd.read_csv(r'Task 5.csv')
```

```python
#assigning variables to DataFrame
StoreTradingDuration = df['Store Trading Duration']
AnnualRevenue = df['Annual Revenue']
```

```python
#creating subplots
fig, ax = plt.subplots(1, 1)

#plotting the line of best fit
z1 = np.polyfit(StoreTradingDuration, AnnualRevenue, 1)
p1 = np.poly1d(z1)
plt.plot(StoreTradingDuration, p1(StoreTradingDuration))

#creating x and y-axis labels and title 
plt.xlabel('Store Trading Duration (Years)')
plt.ylabel('Annual Revenue')
plt.title('Store Trading Duration v Annual Revenue')

#plotting scatter
plt.scatter(StoreTradingDuration, AnnualRevenue, s=0.5)
#creating texts on graph
plt.text(20, 210000,'R = '+ str(round(StoreTradingDuration.corr(AnnualRevenue),3)))

#adjusting height and width of graph
fig.set_figheight(5)
fig.set_figwidth(17)
```
